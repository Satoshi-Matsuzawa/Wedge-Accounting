```python
# This memo visualizes the distribution of urban and rural populations
# This is for Tokyo metro area
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df_tokyo = pd.read_csv('../../Data/Downloaded/urban_rural_pop_tokyo.csv', index_col = ['year_jpn', 'year_wst'] )
```

```python
df_kanagawa = pd.read_csv('../../Data/Downloaded/urban_rural_pop_kanagawa.csv', index_col = ['year_jpn', 'year_wst'])
df_saitama = pd.read_csv('../../Data/Downloaded/urban_rural_pop_saitama.csv',index_col = ['year_jpn', 'year_wst'] )
df_chiba = pd.read_csv('../../Data/Downloaded/urban_rural_pop_chiba.csv',index_col = ['year_jpn', 'year_wst'] )
```

```python
df = df_tokyo + df_kanagawa + df_saitama + df_chiba
```

```python
df = df.reset_index()
```

```python
df['urban_share'] = df['pop_urban'] / df['pop_tot']
```

```python
df['rural_share'] = df['pop_rural'] / df['pop_tot']
```

```python
df
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#plt.rc('pgf', texsystem = 'xelatex')
df.plot(x = 'year_wst', y = 'urban_share', 
#               label = r'Normalized cap wedge $\biggl(\frac{1 + \tau_{R,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               #linestyle = '--',
               color = 'b',
               ax = ax)

#ax.set_rasterized(True)
ax.legend_.remove()
#ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title('The proportion of urban population (nationwide)')
plt.savefig('../../Notes/Facts/urban_share_tokyo.pgf')
plt.show()
```
