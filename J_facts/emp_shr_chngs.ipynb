```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv('../../Data/Cleaned_Old/comb_emp.csv')
```

```python
df['prm_emp_shr_diff'] = df['prm_emp_shr'].diff()
```

```python
df['non_prm_emp_shr_diff'] = df['non_prm_emp_shr'].diff()
```

```python
# create moving average
# create 5 year average
window_5 = 5
```

```python
df['prm_emp_shr_diff_5yr_mean'] = df.prm_emp_shr_diff.rolling(window_5, min_periods = 1).mean()
```

```python
df['non_prm_emp_shr_diff_5yr_mean'] = df.non_prm_emp_shr_diff.rolling(window_5, min_periods = 1).mean()
```

```python
# create moving average
# create 5 year average
window_10 = 10
```

```python
df['prm_emp_shr_diff_10yr_mean'] = df.prm_emp_shr_diff.rolling(window_10, min_periods = 1).mean()
```

```python
df['non_prm_emp_shr_diff_10yr_mean'] = df.non_prm_emp_shr_diff.rolling(window_10, min_periods = 1).mean()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'prm_emp_shr_diff', ax = ax)
plt.title('Primary industry employment diff')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'prm_emp_shr_diff_5yr_mean', ax = ax)
plt.title('Primary industry employment diff 5 year mean')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'prm_emp_shr_diff_10yr_mean', ax = ax)
plt.title('Primary industry employment diff 10 year mean')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'prm_emp_shr', ax = ax)
plt.title('Primary industry employment share')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'non_prm_emp_shr_diff', ax = ax)
plt.title('Non primary industry employment diff')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'non_prm_emp_shr_diff_5yr_mean', ax = ax)
plt.title('Non primary industry employment diff 5 year mean')
plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(x = 'year_wst', y = 'non_prm_emp_shr_diff_10yr_mean', ax = ax)
plt.title('Non primary industry employment diff 10 year mean')
plt.show()
```
