```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv('../../Data/Downloaded/urban_rural_area.csv')
```

```python
df['urban_share'] = df['area_urban'] / df['area_tot']
```

```python
df['rural_share'] = df['area_rural'] / df['area_tot']
```

```python
df
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#plt.rc('pgf', texsystem = 'xelatex')
df.plot(x = 'year_wst', y = 'urban_share', 
#               label = r'Normalized cap wedge $\biggl(\frac{1 + \tau_{R,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               #linestyle = '--',
               color = 'b',
               ax = ax)

#ax.set_rasterized(True)
ax.legend_.remove()
#ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title('The proportion of urban areas (nationwide)')
plt.savefig('../../Notes/Facts/urban_share_area.pgf')
plt.show()
```
