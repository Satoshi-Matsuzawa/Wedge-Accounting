```python
# This memo visualizes the distribution of urban and rural populations
# This is for Osaka metro area
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df_osaka = pd.read_csv('../../Data/Downloaded_Old/urban_rural_pop_osaka.csv', index_col = ['year_jpn', 'year_wst'] )
```

```python
df_kyoto = pd.read_csv('../../Data/Downloaded_Old/urban_rural_pop_kyoto.csv', index_col = ['year_jpn', 'year_wst'])
df_hyogo = pd.read_csv('../../Data/Downloaded_Old/urban_rural_pop_hyogo.csv',index_col = ['year_jpn', 'year_wst'] )
```

```python
df = df_osaka + df_kyoto + df_hyogo
```

```python
df = df.reset_index()
```

```python
df['urban_share'] = df['pop_urban'] / df['pop_tot']
```

```python
df['rural_share'] = df['pop_rural'] / df['pop_tot']
```

```python
df
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#plt.rc('pgf', texsystem = 'xelatex')
df.plot(x = 'year_wst', y = 'urban_share', 
#               label = r'Normalized cap wedge $\biggl(\frac{1 + \tau_{R,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               #linestyle = '--',
               color = 'b',
               ax = ax)

#ax.set_rasterized(True)
ax.legend_.remove()
#ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title('The proportion of urban population (osaka)')
plt.savefig('../../Notes/Facts/urban_share_osaka.pgf')
plt.show()
```
