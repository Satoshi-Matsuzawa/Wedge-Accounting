```python
df_static = df_exp.copy()
```

```python
df_static['food-gammaA'] = df_static.food - gammaA
df_static['cA-gammaA/cM'] = df_static['food-gammaA'] / df_exp.non_food
df_static['muM/muA'] = etaM/etaA * df_static['cA-gammaA/cM']
```

```python
df = df_exp.merge(df_CPI, on = 'year_wst').copy()
```

```python
df['non_food_forward'] = df['non_food'].shift(-1)
```

```python
df['price_forward'] = df['cpi_ind'].shift(-1)
```

```python
df['PtcMt/Pt+1cMt+1'] = df['non_food'] / df['non_food_forward']
```

```python
df['Pt+1/Pt'] = df['price_forward'] / df['cpi_ind']
```

```python
df['cMt/cMt+1'] = df['PtcMt/Pt+1cMt+1'] * df['Pt+1/Pt']
```
