```python
import pandas as pd
import numpy as np
```

```python
# read csv files
df_exp = pd.read_csv('../../Data/Downloads/Post/post_exp.csv')
df_CPI = pd.read_csv('../../Data/Downloads/Post/post_CPI_alt.csv')
```

```python
# Convert CPIs so that 1990 price is 100
```

```python
cpi_goods_1990 = df_CPI.loc[df_CPI.year_wst == 1990]['cpi_goods'].values[0]
```

```python
cpi_agr_1990 = df_CPI.loc[df_CPI.year_wst == 1990]['cpi_agr'].values[0]
```

```python
df_CPI['cpi_goods_1990'] = df_CPI['cpi_goods'] / cpi_goods_1990
df_CPI['cpi_agr_1990'] = df_CPI['cpi_agr'] / cpi_agr_1990
```

```python
df_CPI
```

```python
df = df_exp[['year_wst', 'cons_pc', 'food_pc']].merge(df_CPI[['year_wst', 'cpi_goods_1990', 'cpi_agr_1990']], on = 'year_wst')
```

```python
df['cons_pc_real'] = df['cons_pc'] / df['cpi_goods_1990']
df['food_pc_real'] = df['food_pc'] / df['cpi_agr_1990']
```

```python
df['non_food_pc_real'] = df['cons_pc_real'] - df['food_pc_real']
```

```python
df_real = df[['year_wst', 'cons_pc_real', 'non_food_pc_real', 'food_pc_real']].copy()
```

```python
df_real.columns = ['year_wst', 'cons_pc', 'non_food_pc', 'food_pc']
```

```python
df_real['cons_pc'] = df_real['cons_pc'] * 12
df_real['non_food_pc'] = df_real['non_food_pc'] * 12
df_real['food_pc'] = df_real['food_pc'] * 12
```

```python
df_real.to_csv('test.csv')
```
