```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_pre = pd.read_csv('../../Data/Results/pre_lab_w.csv')
df_post = pd.read_csv('../../Data/Results/post_lab_w.csv')
```

```python
df_merged = pd.concat([df_pre, df_post])
```

```python
df_merged
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.rc('pgf', texsystem = 'xelatex')
df_pre.plot(x = 'year_wst', y = 'norm_labor_wedge-1', 
               label = r'Normalized labor wedge $\biggl(\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               linestyle = '--',
               color = 'b',
               ax = ax)
df_post.plot(x = 'year_wst', y = 'norm_labor_wedge-1',  
               marker = '.',
               linestyle = '--',
               color = 'b',
               ax = ax)
#ax.set_rasterized(True)
ax.legend_.remove()
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title(r'\textbf{Normalized labor wedge $\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1$}')
plt.savefig('../../Notes/comb_lab_w.pgf')
plt.show()
```
