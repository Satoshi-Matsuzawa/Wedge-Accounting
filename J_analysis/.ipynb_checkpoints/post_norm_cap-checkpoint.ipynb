```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv files
df_emp = pd.read_csv('../../Data/Downloaded/postwar_emp.csv')
df_GDP = pd.read_csv('../../Data/Downloaded/postwar_GDP.csv')
df_cns = pd.read_csv('../../Data/Downloaded/postwar_real_cons.csv')
```

```python
# parameters 
α_KA = 0.15
α_LA = 0.55
α_KM = 0.6
α_LM = 0.4
η    = 0.15
γ    = 76.4
```

```python
# create the ratio of marginal utility
df_cns['cA-γ/cM'] = (df_cns['real_food_cons'] - γ) / df_cns['real_non_food_cons']
df_cns['mu_M/mu_A'] = (1-η) / η * df_cns['cA-γ/cM']
```

```python
df_GDP_emp = pd.merge(df_GDP, df_emp, on = 'year_wst')
```

```python
df_GDP_emp_picked = df_GDP_emp[['year_wst', 'prm_GDP', 'non_prm_GDP', 'prm_emp', 'non_prm_emp']]
```

```python
df_GDP_emp_picked['prm_labor_prod'] = α_LA * df_GDP_emp_picked['prm_GDP'] / df_GDP_emp_picked['prm_emp'] 
```

```python
df_GDP_emp_picked['non_prim_labor_prod'] = α_LM * df_GDP_emp_picked['non_prm_GDP'] / df_GDP_emp_picked['non_prm_emp'] 
```

```python
df_GDP_emp_picked
```

```python
# get the ratio of labor productivity
df_GDP_emp_picked['non_prim_labor_prod/prim_labor_prod'] = df_GDP_emp_picked['non_prm_labor_prod'] / df_GDP_emp_picked['prm_labor_prod']
```

```python
df_merged = pd.merge(df_GDP_emp_picked[['year_wst', 'non_prm_labor_prod/prm_labor_prod']], 
                    df_cns[['year_wst', 'mu_M/mu_A']], on = 'year_wst')
```

```python
df_merged['norm_labor_wedge'] = df_merged['mu_M/mu_A'] * df_merged['non_prm_labor_prod/prm_labor_prod']
```

```python
df_merged['norm_labor_wedge-1'] = df_merged['norm_labor_wedge'] - 1
```

```python
df_merged
```

```python
fig, ax = plt.subplots()
df_merged.plot(x = 'year_wst', y = 'norm_labor_wedge-1', ax = ax)
```
