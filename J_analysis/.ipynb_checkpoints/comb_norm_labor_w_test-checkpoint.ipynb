```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_pre = pd.read_csv('../../Data/Results/pre_labor_w.csv')
df_post = pd.read_csv('../../Data/Results/post_labor_w.csv')
```

```python
df_merged
```

```python
df_merged = pd.concat([df_pre, df_post])
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.rc('pgf', texsystem = 'xelatex')
df_pre.plot(x = 'year_wst', y = 'labor_wedge', 
               label = r'Normalized labor wedge $\biggl(\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               linestyle = '--',
               color = 'b',
               ax = ax)
df_post.plot(x = 'year_wst', y = 'labor_wedge',  
               marker = '.',
               linestyle = '--',
               color = 'b',
               ax = ax)
#ax.set_rasterized(True)
ax.legend_.remove()
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title(r'\textbf{Normalized labor wedge $\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1$}')
plt.savefig('../../Notes/comb_lab_w.pgf')
plt.show()
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#plt.rc('pgf', texsystem = 'xelatex')
df_pre.plot(x = 'year_wst', y = 'YM/YA', 
               label = r'Normalized labor wedge $\biggl(\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               linestyle = '--',
               color = 'b',
               ax = ax)

df_pre.plot(x = 'year_wst', y = 'LM/LA', 
               label = r'Normalized labor wedge $\biggl(\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               marker = '.',
               linestyle = '--',
               color = 'c',
               ax = ax)

df_post[df_post.year_wst <= 1975].plot(x = 'year_wst', y = 'YM/YA',  
               marker = '.',
               linestyle = '--',
               color = 'b',
               ax = ax)

df_post[df_post.year_wst <= 1975].plot(x = 'year_wst', y = 'LM/LA',  
               marker = '.',
               linestyle = '--',
               color = 'c',
               ax = ax)
#ax.set_rasterized(True)
ax.legend_.remove()
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title(r'\textbf{Normalized labor wedge $\tau_{W,t}$}')
plt.savefig('../../Notes/comb_lab_w.pdf')
plt.show()
```
