```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_pre = pd.read_csv('../../Data/Results/pre_labor_w.csv')
df_post = pd.read_csv('../../Data/Results/post_labor_w.csv')
```

```python
df_pre['non_prm_emp_2'] = df_pre.tot_emp - df_pre.prm_emp
df_post['non_prm_emp_2'] = df_post.tot_emp - df_post.prm_emp
```

```python
df_pre['LM/LA_2'] = df_pre['non_prm_emp_2'] / df_pre['prm_emp']
```

```python
df_post['LM/LA_2'] = df_post['non_prm_emp_2'] / df_post['prm_emp']
```

```python
df_merged = pd.concat([df_pre, df_post])
```

```python
df_merged.columns
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.rc('pgf', texsystem = 'xelatex')
df_pre.plot(x = 'year_wst', y = 'labor_wedge', 
               label = r'Normalized labor wedge $\biggl(\frac{1 + \tau_{W,t}}{1 + \tau_{C,t}}-1 \biggr)$', 
               linestyle = '--',
               color = 'b',
               ax = ax)
df_post.plot(x = 'year_wst', y = 'labor_wedge',  
               linestyle = '--',
               color = 'b',
               ax = ax)
#ax.set_rasterized(True)
ax.legend_.remove()
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title(r'\textbf{Labor wedge $\tau_{W}$}')
plt.savefig('../../Notes/Discussion/labor_w.pgf')
plt.show()
```
