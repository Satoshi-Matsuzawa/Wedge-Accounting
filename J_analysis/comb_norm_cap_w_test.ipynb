```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
etaA = 0.15
etaM = 0.85
```

```python
df_pre = pd.read_csv('../../Data/Results/pre_cap_w.csv')
df_post = pd.read_csv('../../Data/Results/post_cap_w.csv')
```

```python
df_merged = pd.concat([df_pre, df_post])
```

```python
df_merged.columns
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.rc('pgf', texsystem = 'xelatex')
df_pre.plot(x = 'year_wst', y = 'cap_wedge', 
               linestyle = '--',
               color = 'b',
               ax = ax)
df_post[df_post.year_wst <= 1990].plot(x = 'year_wst', y = 'cap_wedge',  
               linestyle = '--',
               color = 'b',
               ax = ax)
#ax.set_rasterized(True)
ax.legend_.remove()
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title(r'\textbf{Capital wedge $\tau_{R}$}')
plt.savefig('../../Notes/Discussion/cap_w.pgf')
#plt.savefig('../../Graphs/pdf/cap_w.pdf')
plt.show()
```

```python
fig, ax = plt.subplots()

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

#plt.rc('pgf', texsystem = 'xelatex')
df_pre.plot(x = 'year_wst', y = 'YM/YA', 
               label = r'$P^{M}Y^{M}/P^{A}Y^{A}$', 
               linestyle = '--',
               color = 'b',
               ax = ax)

df_pre.plot(x = 'year_wst', y = 'KM/KA', 
               label = r'$K^M/K^A$', 
               linestyle = '-.',
               color = 'c',
               ax = ax)

df_post[df_post.year_wst <= 1990].plot(x = 'year_wst', y = 'YM/YA',  
               label = '_nolegend_',
               linestyle = '--',
               color = 'b',
               ax = ax)

df_post[df_post.year_wst <= 1990].plot(x = 'year_wst', y = 'KM/KA',  
               label = '_nolegend_',
               linestyle = '-.',
               color = 'c',
               ax = ax)
#ax.set_rasterized(True)
#ax.legend_.remove()
plt.xlabel('year')
fig.set_size_inches(w=6.26894, h=3.5)
plt.title(r'\textbf{Capital ratio $K^M/K^A$ and output ratio $P^{M}Y^{M}/P^{A}Y^{A}$}')
plt.savefig('../../Notes/Discussion/output_capital_ratio.pgf')
plt.savefig('../../Graphs/pdf/output_cap_ratio.pdf')
plt.show()
```
