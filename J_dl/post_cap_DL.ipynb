```python
# remove columns and rows with too many nulls
df_dropped = df_original.loc[df_original.notnull().sum(axis=1) > 2, df_original.notnull().sum() > 2]
```

```python
df_dropped
```

```python
# pick total capital and capital in the primary sector
df_picked = df_dropped.iloc[2:, [0, 1,2]]
```

```python
# set column names
df_picked.columns = ['year_quarter', 'tot_cap', 'prm_cap']
```

```python
# select first-quarter data
df_first_quarter = df_picked.iloc[::4]
```

```python
# remove the last six letters of the year column to get year in Japanese calender
df_first_quarter['year_jpn'] = df_first_quarter['year_quarter'].astype(str).str[:-6]
```

```python
df_cleaned = df_first_quarter.drop(['year_quarter'], axis = 1)
```

```python
df_cleaned['non_prm_cap'] = df_cleaned['tot_cap'] - df_cleaned['prm_cap']
```

```python
# swap columns
columns_titles = ["year_jpn","tot_cap", "prm_cap","non_prm_cap"]
df_cleaned=df_cleaned.reindex(columns=columns_titles)
```

```python
df_cleaned['year_jpn'] = df_cleaned['year_jpn'].astype(str)
```
