```python
import pandas as pd
```

```python
df_consumption = pd.read_csv('../../Data/Downloaded/pre_tot_cns.csv')
df_pop = pd.read_csv('../../Data/Downloaded/pre_pop.csv')
```

```python
df_pop
```

```python
df_merged = pd.merge(df_consumption, df_pop, on = 'year_wst')
```

```python
df_merged['food_cns'] = df_merged['food'] / df_merged['tot_pop'] * 1000
# unit is yen at 1934 to 1936 prices 
```

```python
df_merged['non_food_cns'] = df_merged['non_food'] / df_merged['tot_pop'] * 1000
# unit is yen at 1934 to 1936 prices
```

```python
df_merged['tot_cns'] = df_merged['total'] / df_merged['tot_pop'] * 1000
# unit is yen at 1934 to 1936 prices
```

```python
df_picked = df_merged[['year_wst', 'tot_cns', 'food_cns', 'non_food_cns']]
```

```python
df_picked.to_csv('../../Data/Downloaded/pre_cns.csv', index = False)
```
