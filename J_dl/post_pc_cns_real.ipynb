```python
import pandas as pd
df_hh_cons_nom = pd.read_csv('../../Data/Downloaded/post_hh_cns_nom.csv')
df_hh_size = pd.read_csv('../../Data/Downloaded/post_hh_size.csv')
df_price = pd.read_csv('../../Data/Downloaded/post_price.csv')
```

```python
df_hh_size
```

```python
# merge household consumption df and household size df
df_pc_cons_nom = pd.merge(df_hh_cons_nom, df_hh_size, left_on = 'year_wst', right_on = 'year_wst')
```

```python
# create per capita nominal expenditure
df_pc_cons_nom['pc_cns_exp_nom'] = df_pc_cons_nom['Consumption expenditures']/df_pc_cons_nom['hh_size']
```

```python
# create per capita nominal food consumption
df_pc_cons_nom['pc_food_nom'] = df_pc_cons_nom['Food']/df_pc_cons_nom['hh_size']
```

```python
# drop columns
df_pc_cons_nom_selected = df_pc_cons_nom[['year_jpn', 'year_wst', 'pc_cns_exp_nom', 'pc_food_nom']]
```

```python
# merge the df with df of the price level
df_pc_cns_real = pd.merge(df_pc_cons_nom_selected, df_price, on = 'year_wst')
```

```python
# create real consumption expenditure using price index for goods
df_pc_cns_real['pc_cns_exp_real'] = df_pc_cns_real['pc_cns_exp_nom'] / df_pc_cns_real['pi_goods']
```

```python
# create real consumption expenditure using price index for foods
df_pc_cns_real['pc_food_real'] = df_pc_cns_real['pc_food_nom'] / df_pc_cns_real['pi_ag']
```

```python
# create real consumpiton of non food products
df_pc_cns_real['pc_non_food_real'] = df_pc_cns_real['pc_cns_exp_real'] - df_pc_cns_real['pc_food_real']
```

```python
df_cleaned = df_pc_cns_real[['year_jpn_x', 'year_wst', 'pc_cns_exp_real', 'pc_food_real', 'pc_non_food_real']]
```

```python
df_cleaned.columns = ['year_jpn', 'year_wst', 'real_tot_cns','real_food_cns', 'real_non_food_cns']
```

```python
df_cleaned.to_csv('../../Data/Downloaded/post_real_cns.csv',index = False)
```
